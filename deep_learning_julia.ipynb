### Deep learning with Julia - classification neural network

```julia
using Flux
using Statistics
using Plots
```

```julia
#funkcje do generowania sztucznych zbiorów danych
function generate_real_data(n)
    x1 = rand(1,n) .- 0.5
    x2 = (x1 .* x1)*3 .+ randn(1,n)*0.1
    return vcat(x1,x2)
end
function generate_fake_data(n)
    θ  = 2*π*rand(1,n)
    r  = rand(1,n)/3
    x1 = @. r*cos(θ)
    x2 = @. r*sin(θ)+0.5
    return vcat(x1,x2)
end
```

```julia
#utowrzenie sztucznych zbiorów danych
train_size = 5000
real = generate_real_data(train_size)
fake = generate_fake_data(train_size)

#wizualizacja
scatter(real[1,1:500],real[2,1:500])
scatter!(fake[1,1:500],fake[2,1:500])
```

### Utworzenie sieci neuronowej

Metoda `Chain()` służy do utworzenia modelu prostej sieci neuronowej łączącej warstwy w łańcuch. Składa się z dwóch warstw gęstych: pierwsza z 2 wejściami i 25 wyjściami aktywowana funkcją liniową ReLU oraz druga - z 25 wejściami i 1 finalnym wyjściem zadeklarowanym z parametrem sigma, który parametryzuje wynik końcowy do przedziału (0, 1).

```julia
function NeuralNetwork()
    return Chain(
            Dense(2, 25,relu),
            Dense(25,1,x->σ.(x))
            )
end
```

```julia
#modelowanie sztucznie wygenerowanych danych
X = hcat(real,fake)
Y = vcat(ones(train_size),zeros(train_size))
data = Flux.Data.DataLoader((X, Y'), batchsize=100,shuffle=true);

#zdefiniowanie modelu, utworzenie funkcji straty i optymalizatora
m = NeuralNetwork()
opt = Descent(0.05)
loss(x, y) = sum(Flux.Losses.binarycrossentropy(m(x), y))
```

```julia
#Training method 1
ps = Flux.params(m)
epochs = 20
for i in 1:epochs
    Flux.train!(loss, ps, data, opt)
end
println(mean(m(real)),mean(m(fake))) # Print model prediction
```

```julia
#Training method 2
m    = NeuralNetwork()
function trainModel!(m,data;epochs=20)
    for epoch = 1:epochs
        for d in data
            gs = gradient(Flux.params(m)) do
                l = loss(d...)
            end
            Flux.update!(opt, Flux.params(m), gs)
        end
    end
    @show mean(m(real)),mean(m(fake))
end
trainModel!(m,data;epochs=20)
```

```julia
#wizualizacja wyników
scatter(real[1,1:100],real[2,1:100],zcolor=m(real)')
scatter!(fake[1,1:100],fake[2,1:100],zcolor=m(fake)',legend=false)
```
