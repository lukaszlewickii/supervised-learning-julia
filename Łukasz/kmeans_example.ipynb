```julia
using Clustering, Plots
using Random
```

```julia
# generate random points

Random.seed!(1)

f1 = rand(100)

f2 = rand(100)
```

```julia
# plot data

p_rand = scatter(f1, f2,
    xlabel = "Feature 1",
    ylabel = "Feature 2",
    title = "Random Data",
    legend = false
)
```

```julia
# pre-process data
# transpozycja macierzy, kaÅ¼dy wiersz to feature a kaÅ¼da kolumna to sample
X = [f1 f2]'
```

```julia
# initialize variables
# k - liczba klastrÃ³w ktÃ³rÄ… chcemy zainicjalizowaÄ‡
k = 5
# maksymalna liczba iteracji
itr = 100
```

```julia
# cluster data using k-means clustering algorithm

Random.seed!(1)

result = kmeans(X, k; maxiter = itr, display = :iter)

a = assignments(result)

c = counts(result)

mu = result.centers
#macierz centroidÃ³w - pierwszy wiersz to wspÃ³Å‚rzÄ™dna x, drugi to wspÃ³Å‚rzÄ™dna y
```

```julia
# plot results

p_kmeans_demo = scatter(f1, f2,
    xlabel = "Feature 1",
    ylabel = "Feature 2",
    title = "k-means Clustering Demo",
    legend = false,
    group = a,
    markersize = 10,
    alpha = 0.7
)
```

```julia
# plot centroids

scatter!(mu[1, :], mu[2, :],
    color = :yellow,
    markersize = 20,
    alpha = 0.7
)
```
